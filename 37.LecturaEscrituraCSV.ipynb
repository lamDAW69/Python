```python
import csv

# Esto lo que hace es leer la información del documentos 'csv' y la mete en un diccionario
# Función para leer el archivo CSV y devolver una lista de diccionarios
def leer_jugadores(archivo):
    jugadores = []
    with open(archivo, newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            print("Leyendo fila:", row)  # 2Impresión de depuración
            # Verificar que las claves 'Nombre' y 'Puntuación' existan
            if 'Nombre' in row and 'Puntuación' in row:
                jugadores.append({"Nombre": row["Nombre"], "Puntuación": int(row["Puntuación"])})
            else:
                print("Error: Claves 'Nombre' o 'Puntuación' no encontradas en la fila")
    return jugadores

# Funcion para guardar la lista de diccionarios en el archivo csv
def guardar_jugadores(archivo, jugadores):
    with open(archivo, 'w', newline= '') as csvfile: 
        campos = ['Nombre', 'Puntuación']
        writes = csv.DictWriter(csvfile, fieldnames=campos)
        writes.writeheader()
        for jugador in jugadores: 
            writes.writerow(jugador)


# Función para mostrar el menú
def mostrar_menu(): 
    print("1) Listar jugadores")
    print("2) Añadir jugador")
    print("3) Sumar puntuación")
    print("4) Eliminar jugador")
    print("0) Salir")

# Function para listar jugadores
def listar_jugadores(jugadores):
    for i, jugador in enumerate(jugadores): 
        print(f"{i+1}. {jugador['Nombre']}: {jugador['Puntuación']}")


# Función para añadir un jugador
def anadir_jugador(jugadores):
   nombre = input("Ingrese el nombre del jugador: ")
   jugadores.append({"Nombre": nombre, "Puntuación": 0.0})

# Function to remove a player from the list
def sumar_puntuacion_jugador(jugadores):
    listar_jugadores(jugadores)
    pos = int(input("Ingrese el número del jugador: ")) - 1
    puntos = float(input("Ingrese la puntuacuón a sumar: "))
    jugadores[pos]["Puntuación"] += puntos
    jugadores.sort(key=lambda x: x["Puntuación"], reverse=True) # Ordenar jugadores por puntuación

# eliminar_jugador
def eliminar_jugador(jugadores):
    listar_jugadores(jugadores)
    pos = int(input("Ingrese el número del jugador: ")) - 1
    jugadores.pop(pos)

def main():
    archivo = 'players.csv'
    jugadores = leer_jugadores(archivo)

    while True:
        mostrar_menu()
        opcion = input("Selecciona una opción: ")

        if opcion == '1':
            listar_jugadores(jugadores)
        elif opcion == '2':
            anadir_jugador(jugadores)
        elif opcion == '3':
            sumar_puntuacion_jugador(jugadores)
        elif opcion == '4':
            eliminar_jugador(jugadores)
        elif opcion == '0':
            guardar_jugadores(archivo, jugadores)
            break
        else:
            print("Opción no válida, por favor selecciona una opción del menú.")

if __name__ == "__main__":
    main()
```
