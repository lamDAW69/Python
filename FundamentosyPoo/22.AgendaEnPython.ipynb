```python
import builtins



agenda = {

    "Jota":685965478,

    "Elena": 698785689,

    "Raúl":874596588

    }



opcion = 1





while opcion != 0:

    builtins.print(f"""

        1) Añadir

        2) Buscar

        3) Borrar

        4) Listar

        0) Salir

    """)



    opcion = int(input("Selecciona una opción: "))

    

    match opcion:

        case 1:

            nombre = input("Nombre: ")

            telefono = input("Telefono: ")

            agenda[nombre] = telefono

            print(f"{nombre} registrado correctamente!")

        case 2:

            cadena = input("Nombre a buscar: ")

            buscar = {nombre: telefono for nombre,telefono in agenda.items() if nombre.startswith(cadena)}

            if buscar:

                print("Nombre encontrado: ")

                for nombre, telefono in buscar.items():

                    print(f"{nombre} -> {telefono}")

            else:

                print("Nombre no encontrado")   

        case 3:

            cadena = input("Nombre a borrar: ")

            if cadena in agenda:

                seguro = input("¿Estás seguro de que lo quiere borrar?")



                if seguro.lower() == "si":

                    del agenda[cadena]

                    print(f"{cadena} borrado correctamente")

                else: 

                    print("No se ha borrado el nombre")

            else:

                print("Nombre no encontrado")   

        case 4:

            for nombre, telefono in agenda.items():

                print(f"{nombre} -> {telefono}")
```
